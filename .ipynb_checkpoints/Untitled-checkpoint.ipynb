# Text Preprocessing

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def clean_text(text):
    if pd.isna(text):
        return ""
    
    # Convert to lowercase
    text = text.lower()
    
    # Remove URLs
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)
    
    # Remove punctuation and numbers
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    
    # Tokenize
    tokens = word_tokenize(text)
    
    # Remove stopwords and lemmatize
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    
    return " ".join(tokens)
```

```python
data['Cleaned_Text'] = data['CommentText'].apply(clean_text)
```

```python
data[['CommentText', 'Cleaned_Text']].head()
```

# Sentiment Analysis Using TextBlob

```python
def get_sentiment(text):
    polarity = TextBlob(text).sentiment.polarity
    if polarity > 0:
        return "Positive"
    elif polarity < 0:
        return "Negative"
    else:
        return "Neutral"

data['predicted_sentiment'] = data['Cleaned_Text'].apply(get_sentiment)
```

```python
data[['CommentText', 'Cleaned_Text', 'predicted_sentiment']].head()
```

```python
def get_polarity(text):
    return TextBlob(text).sentiment.polarity

data['Sentiment_Score'] = data['Cleaned_Text'].apply(get_polarity)
```

```python
data[['CommentText', 'Sentiment_Score']].head() 
```

# Sentiment Distribution

```python
plt.figure(figsize=(6,4))
sns.countplot(x='predicted_sentiment', data=data, palette='viridis')
plt.title("Sentiment Distribution")
plt.show()
```

```python
positive_text = " ".join(data[data['Sentiment'] == 'Positive']['Cleaned_Text'])
negative_text = " ".join(data[data['Sentiment'] == 'Negative']['Cleaned_Text'])

# Positive WordCloud
wc_pos = WordCloud(width=800, height=400, background_color="white").generate(positive_text)
plt.figure(figsize=(10,5))
plt.imshow(wc_pos.to_image())
plt.title("Positive Sentiment WordCloud")
plt.axis("off")
plt.show()
```

```python
wc_neg = WordCloud(width=800, height=400, background_color="black").generate(negative_text)
plt.figure(figsize=(10,5))
plt.imshow(wc_neg.to_image())
plt.title("Negative Sentiment WordCloud")
plt.axis("off")
plt.show()
```

# Sentiment Over Time (Trend Analysis)

```python
data['PublishedAt'] = pd.to_datetime(data['PublishedAt'])

trend = data.groupby(data['PublishedAt'].dt.date)['predicted_sentiment'] \
            .value_counts().unstack()

trend.plot(figsize=(12,6))
plt.title("Sentiment Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Comments")
plt.legend(title="Sentiment")
plt.show()
```
